##***1. PyKoSpacing***
###***문장 띄어쓰기에 용이***

```python
#@title ***install package*** { display-mode: "form" }
!pip install git+https://github.com/haven-jeon/PyKoSpacing.git
```

```python
sent = '이낙연 캠프 공동선대위원장인 설훈 더불어민주당 의원은 대장동 게이트가 국민의힘 책임 이라는데 동의하시는 국민들보다는 이재명 지사의 책임 이라는데 동의하시는 국민들께서 더 많다고 재차 이재명 후보를 저격했다.'
new_sent = sent.replace(" ", '') # 띄어쓰기가 없는 문장 임의로 만들기
print(new_sent)
```

```python
from pykospacing import Spacing
spacing = Spacing()
kospacing_sent = spacing(new_sent) 

print(sent) # 원래 문장
print(kospacing_sent) # 띄어쓰기가 없는 문장을 띄어쓰기 후 출력
```

##***2. KSS***
###***PykoSpacing 보다 상대적으로 빠르고 문장 분리에 효과적***

```python
#@title ***install package*** { display-mode: "form" }
!pip install kss
```

```python
import kss
s = "이낙연 캠프 공동선대위원장인 설훈 더불어민주당 의원은 대장동 게이트가 국민의힘 책임 이라는데 동의하시는 국민들보다는 이재명 지사의 책임 이라는데 동의하시는 국민들께서 더 많다고 재차 이재명 후보를 저격했다."
for sent in kss.split_sentences(s):
  print(sent)
```
